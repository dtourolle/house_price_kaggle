```python
numerical = {}
cat={}
for column in data:

    convert=[]
    fail = 0
    for d in data[column]:

        try:
            d = float(d)
        except:
            fail+=1
        convert.append(d)
    if fail/len(convert) > 0.9:
        cat[column]=convert
    else:
        numerical[column]=convert
        for i in range(len(numerical[column])):
            try:
                _=float(numerical[column][i])
            except:
                numerical[column][i]=0

```

```python
for name,vals in numerical.items():
    print(name,np.corrcoef(np.array(data['SalePrice'].astype(float)),vals)[0,1])
```

```python
keys={}
for c in cat:

    keys[c]=data[c].unique()
```

```python
cat['BsmtExposure']
```

```python
y = pd.get_dummies(data['HeatingQC'])
print(y.head())
```

```python
one_hoted={}
for k in keys:
    temp = pd.get_dummies(data[k])
    one_hoted[k]={k+'_'+col:temp[col].values for col in temp }
        
```

```python
one_hoted
```

```python
complete_dataset = {}
for n in numerical:
    if n != 'SalePrice' and n != 'Id':
        complete_dataset[n] = np.array(numerical[n])
```

```python
for n,o in one_hoted.items():
    for d,v in o.items():
        complete_dataset[d]=np.array(v)
```

```python
x=np.array(list(complete_dataset.values()))
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(x.transpose(), np.array(numerical['SalePrice']), test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 500)
```

```python
rf.fit(train_features, train_labels)
```

```python
from sklearn.metrics import mean_squared_log_error

# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
# Calculate the absolute errors
errors = mean_squared_log_error(predictions , test_labels)
# Print out the mean absolute error (mae)
print(errors)
```
